# **Implementación de Red Neuronal desde Cero (Problema XOR)**

# **1. Fundamentos Matemáticos de una Red Neuronal Backpropagation**

El objetivo de una red neuronal es aprender una función que mapee entradas a salidas mediante el ajuste de pesos.

## **1.1 Forward Pass (Propagación hacia adelante)**
Es el proceso de calcular la salida de la red. Para cada capa:Se calcula la suma ponderada: $Z = W \cdot X + b$. Se aplica una función de activación (ej. Sigmoide): $A = \sigma(Z)$

- Donde:$W$: Pesos (weights).

- $b$: Sesgo (bias).


- $\sigma(x) = \frac{1}{1 + e^{-x}}$


## **1.2 Backwards Pass (Propagación hacia atrás)**
Utilizamos el descenso del gradiente y la regla de la cadena para encontrar cuánto contribuyó cada peso al error total.

- Calculamos el error en la salida: $E = Y_{real} - Y_{pred}$.

Calculamos la derivada de la función de activación para ajustar los pesos en la dirección que minimice el error.

# **2. Descripción del problema de clasificación por enfrentar**
 El problema XOR (Exclusive OR) es un clásico en el aprendizaje automático. Es una función lógica que devuelve 1 solo si las entradas son diferentes, y 0 si son iguales.

 (0,0) $\rightarrow$ 0

 (0,1) $\rightarrow$

 1(1,0) $\rightarrow$ 1

 (1,1) $\rightarrow$ 0

# **3. Setting up Red Neuronal**

Para resolver el XOR, necesitamos una arquitectura que pueda manejar relaciones no lineales:

Capa de Entrada: 2 neuronas (una para cada bit).

Capa Oculta: 2 a 3 neuronas con activación Sigmoide.

Capa de Salida: 1 neurona (resultado de la clasificación).

Parámetros: * Tasa de aprendizaje (Learning Rate): 0.1

Épocas: 10,000

# **4. Descripción de la librería NumPy**
NumPy es la librería fundamental para computación científica en Python. Permite manejar arreglos multidimensionales y realizar operaciones matemáticas complejas de forma eficiente.

## **3.1 Funciones empleadas**
np.array(): Para crear las matrices de datos.

np.dot(): Para realizar el producto punto entre matrices (pesos y entradas).

np.exp(): Para calcular la función exponencial necesaria en la Sigmoide.

np.random.uniform(): Para inicializar los pesos de forma aleatoria.

## **3.2 Parámetros**
Los arreglos de NumPy se definen por su shape (forma). En este modelo, los pesos tendrán formas de (2, 2) para la capa oculta y (2, 1) para la de salida.

# **5. Pipeline**
## **5.1 Model Selection**
¿Por qué una Red Neuronal para XOR?
Un Perceptrón simple (una sola capa) solo puede resolver problemas linealmente separables. El XOR no se puede separar con una sola línea recta en un plano. Por lo tanto, se requiere una Red Neuronal Multicapa (MLP) que, gracias a sus capas ocultas y funciones de activación no lineales, puede crear fronteras de decisión curvas o complejas.

## **5.2 Model Training**


```python
import numpy as np

# Funciones de activación
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Datos de entrada (XOR)
inputs = np.array([[0,0], [0,1], [1,0], [1,1]])
expected_output = np.array([[0], [1], [1], [0]])

# Inicialización de parámetros
inputLayerNeurons, hiddenLayerNeurons, outputLayerNeurons = 2, 2, 1
lr = 0.1

# Pesos y sesgos
hidden_weights = np.random.uniform(size=(inputLayerNeurons, hiddenLayerNeurons))
hidden_bias = np.random.uniform(size=(1, hiddenLayerNeurons))
output_weights = np.random.uniform(size=(hiddenLayerNeurons, outputLayerNeurons))
output_bias = np.random.uniform(size=(1, outputLayerNeurons))

# Entrenamiento
for _ in range(10000):
    # Forward Pass
    hidden_layer_activation = np.dot(inputs, hidden_weights) + hidden_bias
    hidden_layer_output = sigmoid(hidden_layer_activation)

    output_layer_activation = np.dot(hidden_layer_output, output_weights) + output_bias
    predicted_output = sigmoid(output_layer_activation)

    # Backpropagation
    error = expected_output - predicted_output
    d_predicted_output = error * sigmoid_derivative(predicted_output)

    error_hidden_layer = d_predicted_output.dot(output_weights.T)
    d_hidden_layer = error_hidden_layer * sigmoid_derivative(hidden_layer_output)

    # Actualización de pesos y sesgos
    output_weights += hidden_layer_output.T.dot(d_predicted_output) * lr
    output_bias += np.sum(d_predicted_output, axis=0, keepdims=True) * lr
    hidden_weights += inputs.T.dot(d_hidden_layer) * lr
    hidden_bias += np.sum(d_hidden_layer, axis=0, keepdims=True) * lr

print("Entrenamiento completado.")
```

## **5.3 Prediction**
Función para comprobar los resultados del modelo entrenado:

```python
def predict(input_data):
    # Pasamos los datos por la red ya entrenada
    hidden_layer_input = np.dot(input_data, hidden_weights) + hidden_bias
    hidden_layer_output = sigmoid(hidden_layer_input)

    output_layer_input = np.dot(hidden_layer_output, output_weights) + output_bias
    output = sigmoid(output_layer_input)
    return output

# Comprobación
print("\nResultados finales:")
for i in inputs:
    res = predict(i)
    print(f"Entrada: {i} - Predicción: {res[0].round()} (Valor real: {res[0]})")
```
